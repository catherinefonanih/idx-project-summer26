```python
# identify the number of rows and columns
print(sold.shape)
print(listing.shape)
```

```python
# missing counts
missing_count = sold.isna().count()
print(missing_count)
```

```python
# identify missing values
missing_prop = round((sold.isna().mean()), 2)
print(missing_prop)
```

```python
# identify which columns to drop
cols = missing_prop[missing_prop>0.90].index
cols
```

```python
# drop columns with missing values over 90%
sold = sold.drop(columns = cols)
sold
```
